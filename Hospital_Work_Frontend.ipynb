```python
from ipywidgets import widgets
import FeatureSelectors as fs
import Hospital_Back_End_Program as hsb
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
import matplotlib.pyplot as plt
```

```python
risk_factors_cg = pd.read_csv (r'risk_factors_cg.csv', index_col=0)
consolidated_data = pd.read_csv(r'consolidated_data.csv', index_col=0)

enc = OrdinalEncoder()
enc.fit(np.asarray(consolidated_data[['Mortality 30 days (after discharged)']]).reshape((-1,1)))
da_enc = enc.transform(np.asarray(consolidated_data[['Mortality 30 days (after discharged)']]).reshape((-1,1)))
y_aux=da_enc
y1 = np.ravel([y_aux[i] for i in range(len(risk_factors_cg.index))])

#0-3 outcomes
y2=consolidated_data[['Final Status (Encoded)']]
ls_features = fs.fs(risk_factors_cg,y2)
```

```python
lb = ['Age (Encoded)', 'hsTnl', 'CRP', 'Creatinine', 'Hb', 'Urea', 'LDH', 'ALT', 'Neutrophil', \
      'CK', 'WBC', 'Lymphocyte', 'Total bilirubin', 'Co-morbidities', 'Ferritin', 'Sex (Encoded)']

lbl0 = widgets.Label("Patient Name")
display(lbl0)
text0 = widgets.Text()
display(text0)

lbl1 = widgets.Label(lb[0])
display(lbl1)
text1 = widgets.Text()
display(text1)

lbl2 = widgets.Label(lb[1])
display(lbl2)
text2 = widgets.Text()
display(text2)

lbl3 = widgets.Label(lb[2])
display(lbl3)
text3 = widgets.Text()
display(text3)

lbl4 = widgets.Label(lb[3])
display(lbl4)
text4 = widgets.Text()
display(text4)

lbl5 = widgets.Label(lb[4])
display(lbl5)
text5 = widgets.Text()
display(text5)

lbl6 = widgets.Label(lb[5])
display(lbl6)
text6 = widgets.Text()
display(text6)

lbl7 = widgets.Label(lb[6])
display(lbl7)
text7 = widgets.Text()
display(text7)

lbl8 = widgets.Label(lb[7])
display(lbl8)
text8 = widgets.Text()
display(text8)

lbl9 = widgets.Label(lb[8])
display(lbl9)
text9 = widgets.Text()
display(text9)

lbl10 = widgets.Label(lb[9])
display(lbl10)
text10 = widgets.Text()
display(text10)

lbl11 = widgets.Label(lb[10])
display(lbl11)
text11 = widgets.Text()
display(text11)

lbl12 = widgets.Label(lb[11])
display(lbl12)
text12 = widgets.Text()
display(text12)

lbl13 = widgets.Label(lb[12])
display(lbl13)
text13 = widgets.Text()
display(text13)

lbl14 = widgets.Label(lb[13]+ "1 if has Heart disease/Asthma/Diabetes, 0 otherwise")
display(lbl14)
text14 = widgets.Text()
display(text14)

lbl15 = widgets.Label(lb[14])
display(lbl15)
text15 = widgets.Text()
display(text15)

lbl16 = widgets.Label(lb[15]+ "0 for F 1 for M")
display(lbl16)
text16 = widgets.Text()
display(text16)

btn = widgets.Button(description='Add')
display(btn)
values = [text1,text2,text3,text4,text5,text6,text7,text8,text9,text10,text11,text12,text13,text14,text15,text16]
def add(b):
    temp = ""
    case = pd.DataFrame(columns = ls_features, index = [risk_factors_cg.index[-1]+1])
    i=0
    for lbl in lb:

        #We manually search for these 3 distinct features to encode based on user input
        if lbl in ["Sex (Encoded)"]:
            case[lbl].iloc[0] = int(values[lb.index(lbl)].value)
            continue
            
        elif lbl in ["Age (Encoded)"]:
            if (0.0 <= int(values[lb.index(lbl)].value) <= 30.0):
                case[lbl].iloc[0] = 1.0
            elif (30.0 < int(values[lb.index(lbl)].value) <= 50.0):
                case[lbl].iloc[0] = 2.0
            elif (50.0 < int(values[lb.index(lbl)].value) <= 70.0):
                case[lbl].iloc[0] = 3.0
            else:
                case[lbl].iloc[0] = 4.0
            continue
            
        elif lbl in ["Co-morbidities"]:
            case[lbl].iloc[0] = int(values[lb.index(lbl)].value)
            continue
        
        #Encode the rest
        temp = values[lb.index(lbl)].value
        if ('L' in temp):
            case[lbl].iloc[0] = -1
        elif ('H' in temp):
            case[lbl].iloc[0] = 1
        elif ('*' in temp):
            case[lbl].iloc[0] = 0
        elif ("nan" in temp):
            case[lbl].iloc[0] = 0
        else:
            case[lbl].iloc[0] = 0
    
    print('Result = ',case)
    rfcg=risk_factors_cg
    rfcg = rfcg.append(case)
    print("JESUS1",fs.rus1.predict(case))
    print("JESUS2",fs.gbc2.predict(case))
    conditions = case[ls_features[0:5]].iloc[0].values.tolist()
    flag = 0
    for j in range(len(hsb.df)):
        flag = 0
        for index in hsb.df[j].index:
            if list(index) == conditions:
                print("SATAN ",hsb.medlist[j]," Treatment",hsb.df[j].loc[index])
                flag = 1
        if flag == 0:
            print("Insufficient data to determine treatment probabilties for", hsb.medlist[j])
    rfcg.to_csv (r'risk_factors_cg1.csv', index="False")

btn.on_click(add)
```
